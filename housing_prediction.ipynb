```python
def int_dropnull(i):
  df[i].fillna(value=df[i].mean(),inplace=True)
```

```python
def obj_dropnull(i):
  df[i].fillna(value=df[i].mode()[0],inplace=True)
```

```python
for i in df:
  if(df[i].dtype=="object"):
   obj_dropnull(i)
  else:
    int_dropnull(i)
```

```python
def corr_features(corr_df,threshold):
  arr=corr_df.to_numpy()
  feat=set()
  for i in range(len(arr)):
    for j in range(len(arr)):
      if(j>i):
        if(arr[i,j]>=threshold or arr[i,j]<=-threshold):
         feat.add(corr_df.columns[j])
  return feat
```

```python
corr_df=df.corr(method="spearman")
feat=corr_features(corr_df,0.75)
print(feat)
```

```python
for i in df:
  if i in feat and i!='SalePrice':
    df.drop([i],axis=1,inplace=True)
    df11.drop([i],axis=1,inplace=True)
print(df.shape)
print(df11.shape)
```
